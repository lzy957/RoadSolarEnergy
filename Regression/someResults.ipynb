```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# 中文和负号的正常显示
# plt.rcParams['font.sans-serif'] = 'Microsoft YaHei'
# plt.rcParams['axes.unicode_minus'] = False

# 使用ggplot的绘图风格
# plt.style.use('ggplot')

# Features = pd.read_csv("CityFeature.csv")
fig, axes = plt.subplots(9, 3, figsize=(16,40), subplot_kw=dict(polar=True))
database = pd.read_csv("/datadrive/lzy/Data/BaseData.csv")
for i in range(0,27):
    cityname = database.iloc[i,0]
    print(cityname)
    a = int(i/3)
    b = i%3  
# 构造数据
#     values = np.array(Features.iloc[i,[3,5,12,13,14,15]])
    values = np.array(Features.iloc[i,[2,4,11,12,13,14]])
    feature = ['Lat', 'Clearness', 'GDP', 'P', 'I', 'V']

    N = len(values)
    for j in range(N):
        if values[j] < 1:
            values[j] = values[j] + 1
        if values[j] == 10:
            values[j] = values[j] - 0.3
    # 设置雷达图的角度，用于平分切开一个圆面
    angles=np.linspace(0, 2*np.pi, N, endpoint=False)

    # 为了使雷达图一圈封闭起来，需要下面的步骤
    values=np.concatenate((values,[values[0]]))
    angles=np.concatenate((angles,[angles[0]]))

    # 绘图
#     fig=plt.figure()
    # 这里一定要设置为极坐标格式
#     ax = fig.add_subplot(111, polar=True)
#     axes[a, b].style.use('ggplot')
    # 绘制折线图
    axes[a, b].plot(angles, values, 'o-', linewidth=1,c = 'cornflowerblue')
    # 填充颜色
    axes[a, b].fill(angles, values, alpha=0.25,c = 'cornflowerblue')
    # 添加每个特征的标签
    axes[a, b].set_thetagrids(angles[:-1] * 180/np.pi, feature)
    # 设置雷达图的范围
    axes[a, b].set_ylim(0,10)
    # 添加标题
    axes[a, b].set_title(cityname)
#     plt.title('活动前后员工状态表现')
    # 添加网格线
    axes[a, b].grid(True)
# 显示图形
fig1 = plt.gcf()
# plt.show()
fig1.savefig('Radar_CFeatures.png', dpi=600) 
plt.show()
```
